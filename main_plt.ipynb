```python
x = ['red','green','blue', 'orange', 'yellow']
y = [12,34,23,34,50]
plt.title("Example Bar chart")
plt.xlabel("item colors")
plt.ylabel("numbers")
plt.bar(x,y)
plt.show()
```

```python
x = ['red','green','blue', 'orange', 'yellow']
y = [12,34,23,34,50]
plt.title("Example Pie chart")
plt.pie(y,labels=x, colors=x, autopct= '%1.1f%%')
plt.show()
```
